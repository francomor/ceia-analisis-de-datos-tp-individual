# Analisis de las variables numericas

### Analisis cantidad de bodegas

```python
print(f"media cant_bodegas: {df['cant_bodegas'].mean()}")
print(f"mediana cant_bodegas: {df['cant_bodegas'].median()}")
print(f"moda cant_bodegas: {df['cant_bodegas'].mode()}")
```

```python
fig, ax = plt.subplots(figsize=(6,4))

mean = df["cant_bodegas"].mean()
std = df["cant_bodegas"].std()

ax.hist(df["cant_bodegas"], bins=30, alpha=0.7)
ax.axvline(df["cant_bodegas"].mode()[0], color='yellow', ls='--', lw=2, label='Moda')
ax.axvline(df["cant_bodegas"].median(), color='red', ls='--', lw=2, label='Mediana')
ax.axvline(df["cant_bodegas"].mean(), color='green', ls='--', lw=2, label='Media')
ax.axvline(mean - 3*std, color='red', linestyle='--', label='Mean - 3*STD')
ax.axvline(mean + 3*std, color='green', linestyle='--', label='Mean + 3*STD')
ax.set_title("Distribucion de la cantidad de bodegas")
ax.legend()

plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=df, y='cant_bodegas', color="lightblue")
plt.show()
```

# Analisis de outliers

```python
# Método basado en desviación estándar
df_outlier = df[['cant_bodegas', 'cant_vinedos', 'cant_fabrica']]
mean = df_outlier.mean()
#std = df.std()
std = df_outlier.apply(lambda x: np.nanstd(x, ddof=1))

outliers_std = (df_outlier < (mean - 3 * std)) | (df_outlier > (mean + 3 * std))
outliers_std
```

```python
Q1 = df_outlier.quantile(0.25)
Q3 = df_outlier.quantile(0.75)

IQR = Q3 - Q1
outliers_iqr = (df_outlier < (Q1 - 1.5 * IQR)) | (df_outlier > (Q3 + 1.5 * IQR))
outliers_iqr
```

```python
# Visualización con scatterplot
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.scatterplot(x=df.index, y=df['cant_bodegas'], hue=outliers_iqr['cant_bodegas'], ax=ax[0])
ax[0].set_title("Outliers en 'cant_bodegas' según IQR")
sns.scatterplot(x=df.index, y=df['cant_bodegas'], hue=outliers_std['cant_bodegas'], ax=ax[1])
ax[1].set_title("Outliers en 'cant_bodegas' según std")
plt.show()
```

```python
# Visualización con scatterplot
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.scatterplot(x=df.index, y=df['cant_vinedos'], hue=outliers_iqr['cant_vinedos'], ax=ax[0])
ax[0].set_title("Outliers en 'cant_vinedos' según IQR")
sns.scatterplot(x=df.index, y=df['cant_vinedos'], hue=outliers_std['cant_vinedos'], ax=ax[1])
ax[1].set_title("Outliers en 'cant_vinedos' según std")
plt.show()
```

```python
# Visualización con scatterplot
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.scatterplot(x=df.index, y=df['cant_fabrica'], hue=outliers_iqr['cant_fabrica'], ax=ax[0])
ax[0].set_title("Outliers en 'cant_fabrica' según IQR")
sns.scatterplot(x=df.index, y=df['cant_fabrica'], hue=outliers_std['cant_fabrica'], ax=ax[1])
ax[1].set_title("Outliers en 'cant_fabrica' según std")
plt.show()
```
